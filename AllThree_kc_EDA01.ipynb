# Introduction

* Exploratory data analysis (EDA) a neccessary step prior to building a predictive model. EDA is the process of investigating the dataset for better understanding of the patterns allows for more informed hypothesis testing and more robust model creation. 

#### About the Dataset 



- The dataset contains housing data, which includes features of the houses and the sale price. The data consists of houses sold between May 2014 and May 2015 in King County, Washington. King County is home to the Washington's most populous city, Seattle. 

- The King County dataset is commonly using in data science and machine learning projects.
  
- The features of the dataset are explained below

######  <div align="center">*The Features of the Dataset*

***

- **id** - Unique identification number given to each home sold
- **Date** - Date the home sold  
- **Price** - The Price of the home sold
- **Bedrooms** - The number of bedrooms in the home
- **Bathrooms** - The number of bathrooms in the home
- **Sqft_living** - The size of the living space in square feet
- **Sqft_lot** - The size of the lot in square feet
- **Floors** - The number of floors/levels in the home
- **Waterfront** - A categorical variable of whether the home is located on the waterfront, 0 is used for house that is not on the waterfront, and 1 means that the home is on the waterfront
- **View** -  A categorical variable giving a ranking of the view from the property of the home on a scale of 0 - 4
- **Condition** - A categorical variable of a ranking of the condition of the home in the home's current state on a scale of 1-5
-  **Grade** - A categorical varialbe that is a classification of construction quality such as materials and workmanship on a scale of 1-13
-  **Sqft_above** - The size of the house apart from the basement in square feet
-  **Sqft_basement** - The size of the basement in square feet
-  **yr_built** - The year the house was built
-  **yr_renovated** - The year the house was renovated
-  **lat** - Latitude Coordinate
-  **long** - Longitiude Coordinate
-  **sqft_living15** - The size of the living space of the nearest 15 neighbors in square feet
-  **sqft_lot15** - The size of the lot of the nearest 15 neighbors in square feet  


***


#### The Goal

- The goal of this project will be to create a model to predict the home price. The purpose of this notebook is to show the first part of EDA of the King County Housing Data, the goal is gain further insights of the data through analysis and  visualizations of the features.

# Plot 1 - Histogram of price

- Plot 1 is a Histogram plot that shows the distribution and spread of houses by price. The graph shows a the majority of the houses are sold for under \\$1,000,000 and a large distribution of the houses sell for around \\$500,000.

- Prior to building a model to predict home price, further analysis is needed to gain more of an insights in how the other features influence the distribution of price. 





## Plot 1 - Python

```python
sns.set_style("whitegrid")
fig = plt.subplots(figsize = (10,8))
ax = sns.histplot(house['price'],bins = 225)
#ax.set_xticks(np.arange(0,8000000,500000))
ax.xaxis.set_major_formatter('${x: ,.0f}')
ax.set_yticks(np.arange(0,ax.get_ylim()[1],500))
plt.title('Distribution of Houses by Price')
plt.xlabel('Price')
plt.ylabel('Number of Houses')
plt.show()
```

## Plot 1 - R

```python
%%R -w 900 -h 500 -r 100 -u px
plot_1 <- homes %>% ggplot(aes(x=price)) + 
                    geom_histogram(bins = 225) +
                    scale_x_continuous(labels = scales::dollar_format()) +
                    labs(x = 'price',
                         y = 'Number of Houses',
                         title = 'Frequency of Houses by Price') +
                    theme(plot.title = element_text(hjust = .5), aspect.ratio = .6)

plot_1
```

## Plot 1 - SAS 

```python
%%SAS sas_session
ods graphics on / width= 10in height = 6in;
title 'Frequency of Houses by Price';
Proc SGPLOT data=house;

histogram price / scale = count binstart = 0 nbins= 225;
yaxis label = 'Number of Houses';
xaxis label = 'Price';
/* xaxis grid   values = ('$0','$2,000,000','$4,000,000', '$6,000,000','$8,000,000'); */

xaxis grid values = (0 2000000 4000000 6000000 8000000)
    valuesdisplay = ('0' '$2,000,000' '$4,000,000' '$6,000,000' '$8,000,000');

yaxis grid;

run;
```

# Plot 2 - Correlation

- Plot 2 shows a correlation matrix showing the linear relationship between the features of the dataset. The value displayed in the heatmaps is the Pearson correlation coefficient, a measure of linear correlation between the two variables on a scale of -1 to 1. 

   
- Certain variables such as the square foot of the home, building grade, number of bathrooms, and number of bedrooms have a strong correlation with price. While condition and the year built are not strongly correlated to price, and zipcode is negatively correlated to price.



## Plot 2 - Python

```python
fig = plt.subplots(figsize = (10,8))
sns.heatmap(house.iloc[:,2:].corr(),annot = True, 
            annot_kws={'size':8} ,fmt = ".2f" ,cmap='coolwarm',
            mask = np.triu(house.iloc[:,2:].corr()))
plt.title('Correlation Matrix')
plt.show()
```

## Plot 2 - R

```python
%%R -w 800 -h 500 -r 100 -u px
plot_2 <- ggcorrplot(cor(homes %>% subset(select = - date), use = 'pairwise.complete.obs'),
                     type = 'lower', lab = TRUE, lab_size = 2, sig.level = 0.01) +
                    theme(text = element_text(size = 7),axis.text.x = element_text(size = 7),
                          axis.text.y = element_text(size = 7)) +
                    labs(title = 'Correlation Matrix') +
                    theme(plot.title = element_text(hjust = .5, size = 12),aspect.ratio = .6)
    
plot_2
```

## Plot 2 - SAS 

```python
%%SAS sas_session

title 'Correlation Matrix';
proc format;
	value CorrRangeShade    -1 -< -0.7 = 'Darkblue'
                        -0.69 <- -0.51 = 'blue'
                         -0.509  <-  -0.11 = 'BIGB'
                        -0.109 - 0.109  = 'white'
                        0.11 <- .509 = 'lightred'
                       .51 <- .69 = 'red'
                       0.7 <- 1 = 'moderatered';
run;

proc corr data=house outp=CorrMatrix (where=(_type_='CORR')) noprint;
run;

proc print data=CorrMatrix(drop=_type_ rename=(_name_=Variable))
		   style(column)={backgroundcolor= CorrRangeShade.} noobs;
run;
```

# Plot 3 - Analysis of Bedrooms


- Plot 3 is a bar graph showing the distribution of the number of bedrooms per house. The 3 bedroom home is the most common with number of bedroom per house with 4 bedrooms being the second most common number of bedrooms per house.

- This information could be import later in creation of the model to exclude houses that are outliers in the dataset in order to build a more accurate and robust model.


## Plot 3 - Python

```python
sns.set_style("whitegrid")
fig = plt.subplots(figsize = (10,8))
ax = sns.countplot(house,x ='bedrooms',color = 'grey')
ax.set_yticks(np.arange(0,ax.get_ylim()[1],500))
plt.title('Distribution of Number of Bedrooms per House')
plt.xlabel('Bedrooms')
plt.ylabel('Number of Houses')
plt.show()
```

## Plot 3 - R

```python
%%R -w 800 -h 500 -r 100 -u px
plot_3 <- homes %>% ggplot(aes(x = bedrooms)) +
          geom_bar() + 
          labs(title = 'Distribution of Number of Bedrooms per House',
              x = 'Bedrooms',
             y = 'Number of Houses') +
          theme(plot.title = element_text(hjust = .5))

plot_3
```

## Plot 3 - SAS 

```python
%%SAS sas_session
ods graphics on / width= 10in height = 6in;
title 'Distribution of Number of Bedrooms per House';
proc sgplot data = house;
vbar bedrooms;
yaxis label = 'Number of Houses';
xaxis label = 'Bedrooms';
xaxis grid;
yaxis grid;
run;
```

# Plot 4 - Analysis of Bathrooms


- Plot 4 and Plot 4.1 both are box and whisker plots. The two plots show the distribution of price for each unique number of bathrooms. Plot 4.1 verifies the results of the he histogram in Plot 1 by showing the majority of the houses are sold for under $1,000,000.


- Plot 4 and 4.1 validates the correlation coefficient of 0.53 for the price-bathroom relationship by showing a general pattern of price increase as bathroom number increases as shown by the median of the boxes. Plot 4 shows there are a large amount of outliers for the most common number of bathrooms, and the  IQR (Interquartile range), showing the spread of the data increases around 3.75 bathrooms. These factors could be hindering the correlation coefficient from being more than 0.53.

## Plot 4 & Plot 4.1 - Python

```python
sns.set_style("darkgrid")
fig = plt.subplots(figsize = (12,8))
ax = sns.boxplot(house,x ='bathrooms',y='price',color = 'grey')
ax.yaxis.set_major_formatter('${x: ,.0f}')
ax.set_yticks(np.arange(0,ax.get_ylim()[1],2000000))
plt.title('Home Price by Number of Bathrooms')
plt.xlabel('Bathrooms')
plt.ylabel('Price')
ax.xaxis.grid(True)
plt.show()
```

```python
sns.set_style("darkgrid")
fig = plt.subplots(figsize = (12,8))
ax = sns.boxplot(house,x ='bathrooms',y='price',color = 'grey')
ax = sns.stripplot(house,x ='bathrooms',y='price',color = 'black',jitter = True)
ax.yaxis.set_major_formatter('${x: ,.0f}')
ax.set_yticks(np.arange(0,ax.get_ylim()[1],2000000))
plt.title('Home Price by Number of Bathrooms')
plt.xlabel('Bathrooms')
plt.ylabel('Price')
ax.xaxis.grid(True)
plt.show()
```

## Plot 4 & Plot 4.1 - R

```python
%%R -w 900 -h 600 -r 100 -u px
bathrooms_factor <- as.factor(homes$bathrooms)

plot_4 <- homes %>% ggplot(aes(x = bathrooms_factor, y = price)) + geom_boxplot() + 
          scale_y_continuous(labels = scales::dollar_format())  + 
          labs(x = 'Bathrooms',
               y = 'Price',
               title = 'Home Price by Number of Bathrooms') +
          theme(plot.title = element_text(hjust = .5), axis.text.x = element_text(size = 6))
            options(repr.plot.width=25, repr.plot.height=8)
            

plot_4
```

```python
%%R -w 900 -h 500 -r 100 -u px
bathrooms_factor <- as.factor(homes$bathrooms)


plot_4_1 <- homes %>% ggplot(aes(x = bathrooms_factor, y = price)) + geom_boxplot() + 
          geom_jitter() + 
          scale_y_continuous(labels = scales::dollar_format())  + 
          labs(x = 'Bathrooms',
               y = 'Price',
               title = 'Home Price by Number of Bathrooms') +
          theme(plot.title = element_text(hjust = .5), axis.text.x = element_text(size = 6))


plot_4_1
```

# Plot 5 & Plot 5.1 - Analysis of Home Size 




- Plot 5 and Plot 5.1 are scatterplots of Price by Home Size. The correlation plot shows that the Home Size is the feature with the largest correlation coefficient of 0.7.

- Plot 5.1 shows a log transformation of the  y-axis of the home price. Log transformation is used to better understand the relationship between two variables by reducing the skewness and decrease the impact of variance in the data. 

- The graph shows a general linear pattern of home price increasing as home size increases. The most density of houses in the scatterplot is below \\$1,000,000, with the spread increasing after \\$1,000,000 and \\$2,000,000. The Price - Home Size relationship shows a pattern exhibited in the histogram in Plot 1.


## Plot 5 - Python

```python
sns.set_style("darkgrid")
fig = plt.subplots(figsize = (10,8))
ax = sns.scatterplot(house,x ='sqft_living',y=house['price'],color = 'black')
ax.yaxis.set_major_formatter('${x: ,.0f}')
ax.set_yticks(np.arange(0,ax.get_ylim()[1],2000000))
plt.title('Price by Home Size')
plt.xlabel('Home Size (Sqft)')
plt.ylabel('Price')
ax.xaxis.grid(True)
plt.show()
```

# Plot 6 - Mapping the Houses

- The dataset includes latitude and longitude data allowing for a map of the houses to be created. 

- Plot 6 shows a map of the houses with the color of the points representing the sell price. The map confims that the majority of the houses sell bellow \\$2,000,000. The map shows that the houses clustered around the water sell for more than \\$2,000,000.


## Plot 6 - Python

```python
sns.set_style("darkgrid")
fig, ax = plt.subplots(figsize = (10,8))
sns.scatterplot(x = house['long'],y = house['lat'],hue= house['price'], 
                     palette='viridis', hue_order=list(np.arange(0,max(house['price']),1000000)[1:]))
plt.title('Sell Price by Location of Home')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
ax.xaxis.grid(True)

handles, labels = ax.get_legend_handles_labels()

ax.legend(handles=handles[::-1],
          labels = list(map("${:,.0f}".format, list(np.arange(0,max(house['price']),1000000)[2:7][::-1]))),
           bbox_to_anchor = (1,.55))
plt.show()
```

## Plot 6 - R

```python
%%R -w 800 -h 500 -r 100 -u px
plot_6 <- homes %>% ggplot(aes(x=long, y = lat, color = price)) + 
                        geom_point() + 
        scale_color_continuous(labels = scales::dollar_format(),type = 'viridis') +
          labs(x = 'Longitude',
               y = 'Latitude',
               title = 'Sell Price by Location of Home')+
          theme(plot.title = element_text(hjust = .5))

plot_6
```

## Plot 6 - SAS 

```python
%%SAS sas_session
ods graphics on / width= 10in height = 6in;

proc sgplot data = house;
    scatter y = lat x=long/ name = 'map' colorresponse = price
    colormodel = ThreeColorRamp;
   

*/ gradlegend "$2,000,000" "$4,000,000" "$6,000,000" / title = 'Price'; */ 

*/ continuouslegend 'map' / title = 'PRICE'; */

title 'Sell Price by Home Location';
yaxis label = 'Latitude';
xaxis label = 'Longitude';

gradlegend 'map' / position  = right  title = 'Price($)' extractscale = scientific; 

run;
```

# Analysis of Waterfront properties

- Plot 7 is a box and whisker plot showing the difference in sell price between houses on the waterfront and non-waterfront.
  
- After the map in Plot 6 showed that areas of the more expensive houses were near the water, it seemed important to visualize the price difference between waterfront and non-waterfront properties.
  
- Plot 7 shows that the median waterfront house has a higher sell price than the median Non-waterfront. The relationship between price and waterfront location is characterized by the low correlation coeffcient of 0.27. This is explained by the waterfront location having a larger IQR, thus a larger spread of values around the median sell price. Furthermore, the plot shows a large quantity of outliers for the non-waterfront location. 

## Plot 7 - Python

```python
sns.set_style("darkgrid")
fig = plt.subplots(figsize = (12,8))
ax = sns.boxplot(house,x ='waterfront',y='price',color = 'grey')
ax.yaxis.set_major_formatter('${x: ,.0f}')
ax.set_yticks(np.arange(0,ax.get_ylim()[1],500000))
ax.set_xticklabels(['Non-Waterfront','Waterfront'])
plt.title('Home Price by Waterfront Location')
plt.xlabel('Waterfront')
plt.ylabel('Price')
ax.xaxis.grid(True)
plt.show()
```

## Plot 7 - R

```python
%%R -w 800 -h 600 -r 100 -u px
plot_7 <- homes %>% ggplot(aes(x=as_factor(waterfront), y = price)) + 
                    geom_boxplot() +
                    scale_y_continuous(labels = scales::dollar_format()) +
                    scale_x_discrete(labels = c('Non-Waterfront','Waterfront'))+
                    labs(x = 'Location',
                         y = 'Price',
                         title = 'Price by Waterfront Location')+
                    theme(plot.title = element_text(hjust = .5))

plot_7
```

## Plot 7 - SAS 

```python
%%SAS sas_session
ods graphics on / width= 10in height = 6in;
title 'Price by Waterfront Location';
proc SGPLOT DATA= house;
vbox price / category = waterfront;

yaxis label = 'Price'  values = (0 2000000 4000000 6000000 8000000) 
valuesformat = dollar8.0;
xaxis label = 'Location' values = (0,1)
valuesdisplay = ('Non-Waterfront' 'Waterfront');

run;
```
